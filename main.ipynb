```python
import pandas as pd

# importar as tabelas
tbBase = pd.read_excel("PlanilhaBase.xlsx")
tbTeste = pd.read_excel("Teste.xlsx")
# display(tbTeste)
# tratar a tabela base
tbBase = tbBase.drop(columns="data")
tbBase = tbBase.dropna()
# tbBase.info()
# display(tbBase)
tbBase = tbBase.drop_duplicates(subset=["descrição"], keep="first")
# display(tbBase["descrição"].value_counts())

# entrar em cada linha de coluna descrição na 'tbTeste'
# entrar em cada linha de coluna descrição na 'tbBase' e conferir se alguma dessa linha bate com a linha selecionada na 'tbTeste'
# se bater as linhas, pegar o valor nas colunas "classificação" e "grupo" na 'tbBase' e colocar na 'tbTeste'
tbTeste["classificação"] = pd.NA
tbTeste["grupo"] = pd.NA

for linhaTeste in tbTeste["descrição"]:
    for linhaBase in tbBase["descrição"]:
        if linhaTeste == linhaBase:
            tbTeste.loc[tbTeste["descrição"] == linhaTeste, "classificação"] = tbBase.loc[tbBase["descrição"] == linhaBase, "classificação"]

display(tbTeste)
```
